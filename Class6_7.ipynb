```python
# reading the data 
import pandas as pd
df_loan_data = pd.read_csv('loans data.csv')
df_loan_data.info()

# Correcting the datatyoes due to dots
incorrect_datatype_cols_with_dots = ['Amount.Requested','Amount.Funded.By.Investors','Open.CREDIT.Lines','Revolving.CREDIT.Balance']
for i in incorrect_datatype_cols_with_dots:
  df_loan_data[i] = pd.to_numeric(df_loan_data[i],errors = 'coerce')

# Correcting the datatypes due to %
incorrect_datatype_cols_with_percentage_symbol = ['Interest.Rate','Debt.To.Income.Ratio']
for i in incorrect_datatype_cols_with_percentage_symbol:
  df_loan_data[i] = df_loan_data[i].str.replace('%','')
  df_loan_data[i] = df_loan_data[i].astype('float64')


# Correcting the FICO Score
fico_split = df_loan_data['FICO.Range'].str.split('-',expand = True)
fico_split.columns = ['lrange','uprange']
fico_split['lrange'] = fico_split['lrange'].astype('float')
fico_split['uprange'] = fico_split['uprange'].astype('float')
fico_split['Average Fico'] = ((fico_split['lrange']) + (fico_split['uprange']))/2
df_loan_data['Average_Fico'] = fico_split['Average Fico']
df_loan_data.drop(['FICO.Range'],axis=1,inplace = True)

# We will identify missing values and treat them 
columnnames_missing_val = list(df_loan_data.isnull().sum()[df_loan_data.isnull().sum()>0].index)
for i in columnnames_missing_val:
  if i=='ID':
    df_loan_data.drop(df_loan_data.loc[df_loan_data[i].isnull(),:].index,axis=0,inplace = True)
  else:
    if df_loan_data[i].dtype == 'float64':
      df_loan_data[i].fillna(df_loan_data[i].median(),inplace = True)
    else:
      df_loan_data[i].fillna(df_loan_data[i].mode()[0],inplace = True)

# We will finally do the dummification
df_loan_data_modified = pd.get_dummies(df_loan_data)
```
